```python
import pyisomme
```

```python
channel = pyisomme.create_sample("1???????????????", t_range=(0,0.1,100), y_range=(0,10), mode="linear", unit="m/(s*s)")
channel.data.plot()
channel.data.head()
```

## Differentiate

```python
channel_diff = channel.differentiate()
channel_diff.data.plot()
channel_diff.data.head()
```

## Integrate

```python
channel_int = channel.integrate()
channel_int.data.plot()
channel_int.data.head()
```
