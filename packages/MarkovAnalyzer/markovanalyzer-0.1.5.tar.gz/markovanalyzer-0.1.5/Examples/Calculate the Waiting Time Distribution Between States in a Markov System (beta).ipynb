```python
from markovanalyzer.polyspectra_calculator import System as MarkovSystem
import numpy as np
import matplotlib.pyplot as plt
```

```python
gamma_SO = 50
gamma_C = 1
gamma_SI = 1

rates = {'0->1': gamma_SO,
         '1->0': gamma_SO,
         '0->2': gamma_C,
         '2->0': gamma_C,
         '0->3': gamma_SO,
         '3->0': gamma_SO,
         '0->4': gamma_C,
         '4->0': gamma_C,
         '1->2': gamma_SI,
         '2->1': gamma_SI,
         '2->3': gamma_SI,
         '3->2': gamma_SI,
         '3->4': gamma_SI,
         '4->3': gamma_SI,
         '4->1': gamma_SI,
         '1->4': gamma_SI,}
         
m_op = np.array([1,0,0,0,0])

markov_system = MarkovSystem(rates, m_op)
```

```python
t = np.linspace(0,5,1000)
down_states = [1,2,3,4]
up_states = [0]

wtd_in, wtd_out = markov_system.calculate_WTD(t, down_states, up_states)
```

```python
plt.plot(t, wtd_in)
plt.yscale('log')
```

```python
plt.plot(t, wtd_out)
plt.yscale('log')
```
